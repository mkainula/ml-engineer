```python
adult1_json.reset_index(level=0, inplace=True)
adult2_json.reset_index(level=0, inplace=True)
```

```python
adult_json_merged = adult1_json.merge(adult2_json, on=['index'], how='outer')
```

```python
adult_json_merged.drop('index', 1, inplace=True)
```

```python
adult_json = adult_json.append(adult_json_merged).drop_duplicates()
```

### XML

```python
from lxml import etree
```

```python
parser = etree.XMLParser(recover=True)
```

```python
with open('adult.xml', 'r') as myfile:
  data = myfile.read()
```

```python
parsed_xml = etree.fromstring(data, parser=parser)
parsed_xml_ET = etree.XML(data, parser=parser)
```

```python
parsed_items = []
for item in iter(parsed_xml_ET):
    item_dict = dict()
    item_dict[item.attrib.get("name")] = item.text
    parsed_items.append(item_dict)
```
