### XML

```python
from lxml import etree
```

```python
with open('adult.xml', 'r') as xml_file:
  fixed_xml = '<items>' + xml_file.read().replace('<=50K', '&lt;=50K').replace('>50K', '&gt;50K').replace('&', '&amp;') + '</items>'
```

```python
parsed_xml = etree.XML(fixed_xml)
```

```python
parsed_items = []
for item in iter(parsed_xml):
    item_dict = dict()
    for field in iter(item):
        item_dict[field.attrib.get("name")] = item.text
    parsed_items.append(item_dict)
```

```python
adult_xml = pd.DataFrame(parsed_items)
```

```python
adult_xml.shape
```

### PostgreSQL

```python
adult_sql = pd.read_sql('SELECT * FROM fact_adult f ' +
                    'INNER JOIN dim_capital cap ON cap.id = f.id_capital ' +
                    'INNER JOIN dim_demografic dem ON dem.id = f.id_demografic ' +
                    'INNER JOIN dim_education edu ON edu.id = f.id_education', 'postgresql://mle:mle@127.0.0.1:5432/mle')
```

```python
adult_sql.head()
```

```python
adult_sql.shape
```

```python
adult_sql = adult_sql.T.drop_duplicates().T
```

```python
adult_sql.shape
```

```python
adult_sql.head()
```

```python
adult_sql.drop(['id_demografic', 'id_education', 'id_capital', 'id'], 1, inplace=True)
```

### Merge data

```python
df = adult_json_merged.append(adult_parquet, sort=False).drop_duplicates()
```

```python
df = df.append(adult_xml, sort=False).drop_duplicates()
```

```python
df = df.append(adult_sql, sort=False).drop_duplicates()
```

```python
df.head()
```

```python
df.shape
```

### Train model

```python
import sklearn.preprocessing as preprocessing
from sklearn.model_selection import train_test_split

encoder = preprocessing.LabelEncoder()
df['sex_dummy'] = encoder.fit_transform(df['sex'])
df['income_dummy'] = encoder.fit_transform(df['class'])
x = pd.get_dummies(df[['age','workclass','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country']])
y = df['income_dummy']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
rf.score(x_test, y_test)
```
