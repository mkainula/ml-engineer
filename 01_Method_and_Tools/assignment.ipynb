# Task 2
create a simple model to "explain" variable "marketvalue" and assume that all lines are coming from same population

```r
forbes_lm1 <- lm(marketvalue ~ profits, data=Forbes2000)
```

```r
summary(forbes_lm1)
```

```r
forbes_lm2 <- lm(marketvalue ~ profits + assets + sales, data = Forbes2000)
```

```r
summary(forbes_lm2)
```

```r
xyplot(marketvalue~profits, data = Forbes2000, t=c("p", "r"))
```
